```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv('/kaggle/input/football-players-data/fifa_players.csv')
df1.info()
```

```python
df1.head()
```

```python
df1['positions'].unique()
```

```python
df1.isnull().sum()
```

```python
df1 = df1.drop(['national_team', 'national_jersey_number', 'national_team_position',
                'national_rating', 'release_clause_euro', 'wage_euro', 'value_euro'], axis=1)
df1.info()
```

```python
def classify_position_minimal(pos_str):
    if pos_str == "GK":
        return None  
    primary = pos_str.split(',')[0]
    
    forward_markers = {"CF", "ST", "FW", "LW", "RW"}
    defender_markers = {"CB", "LB", "RB", "LWB", "RWB"}
    midfielder_markers = {"CM", "CAM", "CDM", "LM", "RM"}
    
    if primary in forward_markers:
        return "FW"
    elif primary in defender_markers:
        return "DF"
    elif primary in midfielder_markers:
        return "MF"
    else:
        return "MF"

df1['positions'] = df1['positions'].apply(classify_position_minimal)
```

```python
df1.head()
```

```python
df1['positions'].unique()
```

```python
df1.columns
```

## FW

```python
fw_features = [
    'finishing',         
    'heading_accuracy',  
    'dribbling',         
    'shot_power',        
    'acceleration',      
    'sprint_speed',      
    'positioning',       
    'reactions',         
    'ball_control',      
    'long_shots',        
    'overall_rating',    
    'potential',         
    'composure'
]
```

```python
fw_df1 = df1[df1["positions"] == "FW"].copy()
fw_df1 = fw_df1[fw_features].dropna()
```

```python
scaler = StandardScaler()
df1_scaled = scaler.fit_transform(fw_df1)

kmeans_fw = KMeans(n_clusters=3, random_state=42, n_init=10)
fw_df1["Cluster"] = kmeans_fw.fit_predict(df1_scaled)
```

```python
print(fw_df1["Cluster"].value_counts())
```

```python
centroids = kmeans_fw.cluster_centers_
```

```python
centroids_original = scaler.inverse_transform(centroids)
print(pd.DataFrame(centroids_original, columns=fw_features))
```

```python
fw_df1.info()
```

```python
fw_df1['composite_score'] = fw_df1.mean(axis=1)

fw_ranked = fw_df1.sort_values(by='composite_score', ascending=False)
fw_df1.head()
```

```python
fw_df1['composite_score'] = fw_df1.mean(axis=1)

cluster_scores = fw_df1.groupby('Cluster')['composite_score'].mean().sort_values(ascending=False)

mapping = {cluster_scores.index[0]: 1,
           cluster_scores.index[1]: 0,
           cluster_scores.index[2]: 2}

fw_df1['ranking'] = fw_df1['Cluster'].map(mapping)
```

```python
fw_df1.head()
```

```python
from sklearn.model_selection import train_test_split

X = fw_df1[fw_features]
y = fw_df1['ranking']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, accuracy_score

xgb_model = xgb.XGBClassifier(objective="multi:softmax", num_class=3, random_state=42)
```

```python
#FW
xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
